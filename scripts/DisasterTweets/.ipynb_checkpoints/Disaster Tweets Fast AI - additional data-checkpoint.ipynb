```python
from fastai.text import *
import pandas as pd
import numpy as np
import re
import nltk
from nltk import word_tokenize
from tqdm import tqdm
```

## Simpler model (no text cleaning) for speed

## Preparing the data

```python
input_path = './kaggle/input/nlp-getting-started/'
output_path = './kaggle/working/'
```

```python
# relabel incorrect tweets
ids_with_target_error = [328,443,513,2619,3640,3900,4342,5781,6552,6554,6570,6701,6702,6729,6861,7226]
train_orig = pd.read_csv(input_path+'train.csv', encoding = 'latin-1')
incorrect = train_orig[train_orig['id'].isin(ids_with_target_error)]
```

```python
df_train = pd.read_csv(input_path+'socialmedia-disaster-tweets-DFE.csv', encoding = 'latin-1')
df_train = df_train[['_unit_id','keyword','location','text','choose_one']]
#now grab ids that were identified earlier as mislabelled
incorrect_ids = incorrect.merge(df_train,on=['text'])['_unit_id']
df_train.rename(columns = {'_unit_id': 'id'}, inplace = True)
df_train['target'] = df_train['choose_one'].map({'Relevant': 1, 'Not Relevant': 0})
df_train.drop(['choose_one'], axis = 1, inplace = True)
df_train = df_train[-df_train['target'].isna()]
df_train = df_train.astype({'target': 'int64'})
#reassign incorrect labels
df_train.loc[df_train['id'].isin(incorrect_ids),'target'] = 0
df_train.head()
```

```python
df_train['text'][1]
```

```python
df_test = pd.read_csv(input_path+'test.csv')
df_test.head()
```

```python
df_train.head()
```

# Cleaning text data

```python
# Additional training data
cols = ['target','ids','date','flag','user','text']
train_add = pd.read_csv(input_path+'training.1600000.processed.noemoticon.csv',names=cols, encoding = 'latin-1')
train_add.head()
```

```python
df = df_test[['text']].append(df_train[['text']])
df = df.drop_duplicates().reset_index()['text']
df.head()
```

## Language Model

```python
# start with something smaller and see how GPU handles it
bs=24
```

```python
data_lm = (TextList.from_df(pd.DataFrame(df), cols='text')
           #Inputs: the train.csv file
            .split_by_rand_pct(0.1)
           #We randomly split and keep 10% (10,000 reviews) for validation
            .label_for_lm()           
           #We want to do a language model so we label accordingly
            .databunch(bs=bs))
data_lm.save(output_path+'data_lm_raw.pkl')
```

```python
data_lm = load_data(output_path, 'data_lm_raw.pkl', bs=bs)
```

```python
data_lm.show_batch()
```

```python
learn = language_model_learner(data_lm, AWD_LSTM, drop_mult=0.3)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot(skip_end=15)
```

```python
learn.fit_one_cycle(1, 1e-2, moms=(0.8,0.7))
```

```python
learn.save('fit_head_raw')
```

```python
learn.load('fit_head_raw');
```

```python
learn.unfreeze()
```

```python
learn.fit_one_cycle(10, 1e-3, moms=(0.8,0.7))
```

```python
learn.save('fine_tuned_raw')
```

```python
learn.load('fine_tuned_raw');
```

```python
TEXT = "Emergency happening in"
N_WORDS = 40
N_SENTENCES = 2
```

```python
print("\n".join(learn.predict(TEXT, N_WORDS, temperature=0.75) for _ in range(N_SENTENCES)))
```

```python
learn.save_encoder('fine_tuned_enc_raw')
```

## Classifier

```python
data_clas = (TextList.from_df(df_train, cols='text', vocab=data_lm.vocab)
           #Inputs: the train.csv file
            .split_by_rand_pct(0.1)
           #We randomly split and keep 10% (10,000 reviews) for validation
            .label_from_df(cols='target')          
           #We want to do a language model so we label accordingly
            .databunch(bs=bs))
data_clas.save(output_path+'data_clas_raw.pkl')
```

```python
data_clas = load_data(output_path, 'data_clas_raw.pkl', bs=bs)
```

```python
data_clas.show_batch()
```

```python
learn = text_classifier_learner(data_clas, AWD_LSTM, drop_mult=0.5)
learn.load_encoder('fine_tuned_enc_raw')
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(1, 2e-2, moms=(0.8,0.7))
```

```python
learn.save('first_raw')
```

```python
learn.load('first_raw');
```

```python
learn.freeze_to(-2)
learn.fit_one_cycle(1, slice(1e-2/(2.6**4),1e-2), moms=(0.8,0.7))
```
