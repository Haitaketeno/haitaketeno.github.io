```python
from fastai.text import *
import pandas as pd
import numpy as np
```

## Preparing the data

```python
input_path = './kaggle/input/nlp-getting-started/'
output_path = './kaggle/working/'
```

```python
df_train = pd.read_csv(input_path+'train.csv')
df_train.head()
```

```python
df_train['text'][1]
```

```python
df_test = pd.read_csv(input_path+'test.csv')
df_test.head()
```

```python
df = df_test[['text']].append(df_train[['text']])
df.head()
```

# Cleaning text data

## Language Model

```python
# start with something smaller and see how GPU handles it
bs=24
```

```python
data_lm = (TextList.from_df(df, cols='text')
           #Inputs: the train.csv file
            .split_by_rand_pct(0.1)
           #We randomly split and keep 10% (10,000 reviews) for validation
            .label_for_lm()           
           #We want to do a language model so we label accordingly
            .databunch(bs=bs))
data_lm.save(output_path+'data_lm.pkl')
```

```python
data_lm = load_data(output_path, 'data_lm.pkl', bs=bs)
```

```python
data_lm.show_batch()
```

```python
learn = language_model_learner(data_lm, AWD_LSTM, drop_mult=0.3)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot(skip_end=15)
```

```python
learn.fit_one_cycle(1, 1e-2, moms=(0.8,0.7))
```

```python
learn.save('fit_head')
```

```python
learn.load('fit_head');
```

```python
learn.unfreeze()
```

```python
learn.fit_one_cycle(10, 1e-3, moms=(0.8,0.7))
```

```python
learn.save('fine_tuned')
```

```python
learn.load('fine_tuned');
```

```python
TEXT = "Emergency happening in"
N_WORDS = 40
N_SENTENCES = 2
```

```python
print("\n".join(learn.predict(TEXT, N_WORDS, temperature=0.75) for _ in range(N_SENTENCES)))
```

```python
learn.save_encoder('fine_tuned_enc')
```

## Classifier

```python
data_clas = (TextList.from_df(df_train, cols='text', vocab=data_lm.vocab)
           #Inputs: the train.csv file
            .split_by_rand_pct(0.1)
           #We randomly split and keep 10% (10,000 reviews) for validation
            .label_from_df(cols='target')          
           #We want to do a language model so we label accordingly
            .databunch(bs=bs))
data_clas.save(output_path+'data_clas.pkl')
```

```python
data_clas = load_data(output_path, 'data_clas.pkl', bs=bs)
```

```python
data_clas.show_batch()
```

```python
learn = text_classifier_learner(data_clas, AWD_LSTM, drop_mult=0.5)
learn.load_encoder('fine_tuned_enc')
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(1, 2e-2, moms=(0.8,0.7))
```

```python
learn.save('first')
```

```python
learn.load('first');
```

```python
learn.freeze_to(-2)
learn.fit_one_cycle(1, slice(1e-2/(2.6**4),1e-2), moms=(0.8,0.7))
```

```python
learn.save('second')
```

```python
learn.load('second');
```

```python
learn.freeze_to(-3)
learn.fit_one_cycle(1, slice(5e-3/(2.6**4),5e-3), moms=(0.8,0.7))
```

```python
learn.save('third')
```

```python
learn.load('third');
```

```python
learn.unfreeze()
learn.fit_one_cycle(2, slice(5e-2/(2.6**4),4e-2), moms=(0.8,0.7))
```

```python
learn.predict("I really loved that movie, it was awesome!")
```

```python
learn.save('fourth')
```

```python
learn.load('third');
```

```python
preds = []
for line in df_test.text:
    p = learn.predict(line)
    preds.append(str(p[0]))
```

```python
df_test['target'] = preds
submission = df_test[['id','target']]
submission.head()
```

```python
submission.to_csv('submission.csv',index=False)
```
